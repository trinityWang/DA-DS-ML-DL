```python
get_student_info_sql = """
select s.id,cc.id,cc.full_mark from 
grade g
join 
class c
on g.id = c.grade_id
join student s
on c.id = s.class_id
join course cc 
on g.id = cc.grade_id
"""

cursor.execute(get_student_info_sql)
# fetch数据
result=cursor.fetchall()
print(result)
```
